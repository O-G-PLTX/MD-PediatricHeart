### Detecção de dados faltantes

```python
def missing_data_report(df):
    missing = df.isnull().sum()
    percent = (missing / len(df)) * 100
    report = pd.DataFrame({'Missing Values': missing, 'Percent (%)': percent})
    return report

# Exemplo de uso:
missing_data_report(df)
```
