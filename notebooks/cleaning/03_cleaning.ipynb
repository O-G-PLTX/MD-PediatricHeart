```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import os
```

```python
from dotenv import load_dotenv
load_dotenv()
CSV_PATH = os.getenv("CSV_PATH")
print(f"Caminho do arquivo CSV: {CSV_PATH}")
```

```python
df = pd.read_csv(CSV_PATH)
df.head()
```

```python
df['SOPRO'].value_counts()
```

```python
df['SOPRO'] = df['SOPRO'].replace({'sistólico': 'Sistólico', 'diastólico': 'Diastólico', 'contínuo': 'Contínuo', 'ausente': 'Ausente'
                                   , 'Sistolico e diastólico': 'Sistólico e Diastólico'})
df['SOPRO'].value_counts()
```

```python
df['SEXO'] = df['SEXO'].replace({'Masculino': 'M', 'masculino': 'M', 'Feminino': 'F', 'feminino': 'F'})
df['SEXO'].value_counts()
```

```python
df = df.drop(columns=['HDA2'])
df.head()
```
